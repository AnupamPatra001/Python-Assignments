1. What is the difference between a function and a method in Python ?

8. What is a lambda function in Python and when is it typically used ?
  - Lambda function is an anonymous function, which has no function name and defined using lambda keyword.

9. Explain the purpose and usage of the `map()` function in Python.
  - The map() function is used to execute a function for each value in an itarable.
  - Syntax: map(function , *iterable)
10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python ?
  - map():
     - It executes a function for each value in an itarable.
     - syntax: map(function, *iterable)
  - reduce():
     - It reduces all the iterable values to one value by performing a function on them.
     - syntax: reduce(function, *iterable)
  - filter():
     - It filters out the values of an iterable with a given condition.
     - syntax: filter(function, *iterable)

```python
#1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def ev_sum(lis):
  sum=0
  for i in lis:
    if i%2==0:
      sum+=i
  return sum
a=ev_sum([4,3,7,1,9,2,4,9])
print("Sum of even numbers is",a)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def ch_pr(a):
  if a > 200 :
    print("Out of range. Range(1-200)")
  else:
    if a <= 1:
      return False
    for i in range(2,int(a**0.5)+1):
      if a%i==0:
        return False
      return True
ch_pr(200)
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a
```

```python

# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.rstrip('\n')

for line in read_file_lines('sample.txt'):
    print(line)
```
